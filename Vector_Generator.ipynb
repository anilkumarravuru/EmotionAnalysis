```python
# Anil Kumar Ravuru
import pandas as pd

df = pd.read_csv('Final.csv')
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
vectorizer = CountVectorizer(ngram_range=(1, 2), min_df=2, max_df=1.0)
transformer = TfidfTransformer(use_idf=True)

tf_idf_values = vectorizer.fit_transform(df['token_words'])
```

```python
tf_idf_values.shape
```

```python
reaction_order = ['fb_love', 'fb_haha', 'fb_wow', 'fb_sad', 'fb_angry']
df['top_reaction'] = df['top_reaction'].apply(lambda x: reaction_order.index(x))
df.head()
```

```python
import scipy.sparse
tf_idf_df = pd.DataFrame.sparse.from_spmatrix(tf_idf_values)
```

```python
tf_idf_df['reaction'] = df['top_reaction']
```
